```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df
```

```python
sns.countplot(df, y='class')
plt.title("Passengers Count by Class")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

```python
sns.histplot(df, x="age", kde=True, color='red')
plt.title("Distribution of Passengers Age")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
sns.lineplot(df, x="class", y='fare', linestyle='--',
              marker='o', color='green')
plt.title("Fare Distribution by Class")
plt.xlabel("Class")
plt.ylabel("Fare")
plt.show()
```
